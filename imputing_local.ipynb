```python
import tarfile
import pandas as pd

def extract_csv_from_tar(tar_path):
    dataframes = []
    with tarfile.open(tar_path, 'r') as tar:
        # Iterate over each member in the tar archive
        for member in tar.getmembers():
            if member.isfile() and member.name.endswith('.csv'):
                # Extract the file object
                file_obj = tar.extractfile(member)
                # Read the CSV file into a DataFrame
                df = pd.read_csv(file_obj)
                # Append the DataFrame to the list
                dataframes.append(df)
    return dataframes

# Example usage
tar_path_train = '/Users/grigorijcebotarev/Desktop/thesis/sent.tar.gz'
all_features_speed_not_null = extract_csv_from_tar(tar_path_train)

tar_path_test = '/Users/grigorijcebotarev/Desktop/thesis/sent.tar_test.gz'
all_features_speed_null = extract_csv_from_tar(tar_path_test)

```

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
from sklearn.metrics import roc_auc_score, accuracy_score, mean_squared_error, mean_absolute_error, r2_score
import lightgbm as lgb
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.linear_model import LogisticRegression

X = np.array(all_features_speed_not_null)
y = np.array(y_data)
cls_map = {0.5: 0, 0.8: 1, 1.0: 2, 1.2: 3, 1.5: 4}
inv_cls_map = {v: k for k, v in cls_map.items()}
y_cls = np.array([cls_map[y_i] for y_i in y])

X_tr, X_te, y_tr, y_te = train_test_split(X, y_cls, test_size=0.2, random_state=42)

sc = StandardScaler()
X_tr_shape = X_tr.shape
X_te_shape = X_te.shape

X_tr_flat = X_tr.reshape(-1, X_tr.shape[-1])
X_te_flat = X_te.reshape(-1, X_te.shape[-1])
sc.fit(X_tr_flat)

X_tr_s = sc.transform(X_tr_flat).reshape(X_tr_shape)
X_te_s = sc.transform(X_te_flat).reshape(X_te_shape)
X_s = sc.transform(X.reshape(-1, X.shape[-1])).reshape(X.shape)

dtr = xgb.DMatrix(X_tr_s.reshape(-1, X.shape[1] * X.shape[2]), label=y_tr)
dte = xgb.DMatrix(X_te_s.reshape(-1, X.shape[1] * X.shape[2]), label=y_te)
dall = xgb.DMatrix(X_s.reshape(-1, X.shape[1] * X.shape[2]))

params = {
    'objective': 'multi:softmax',
    'num_class': 5,
    'max_depth': 6,
    'eta': 0.1,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'eval_metric': 'mlogloss'
}

rounds = 100
bst = xgb.train(params, dtr, rounds, evals=[(dte, 'test')], early_stopping_rounds=10)

pred_te = bst.predict(dte).astype(int)
print(f'predicted classes test {pred_te}')

acc = accuracy_score(y_te, pred_te)
print(f'accuracy: {acc}')
mse = mean_squared_error(y_te, pred_te)
print(f'mse: {mse}')
mae = mean_absolute_error(y_te, pred_te)
print(f'mae: {mae}')
r2 = r2_score(y_te, pred_te)
print(f'r2_score: {r2}')

precision = precision_score(y_te, pred_te, average='weighted')
recall = recall_score(y_te, pred_te, average='weighted')
f1 = f1_score(y_te, pred_te, average='weighted')

print(classification_report(y_te, pred_te, target_names=[str(k) for k in cls_map.keys()]))

pred_all = bst.predict(dall).astype(int)
print(f'predicted classes all {pred_all}')

pred_val_te = np.array([inv_cls_map[cls] for cls in pred_te])
pred_val_all = np.array([inv_cls_map[cls] for cls in pred_all])

print(f'predicted test {pred_val_te}')
print(f'predicted all {pred_val_all}')
```
